#Advanced Visuals in Power Bi

1. What is the difference between a Basic Filter, Advanced Filter, and a Slicer in Power BI?
- | Feature | Basic Filter             | Advanced Filter              | Slicer         |
| ------- | ------------------------ | ---------------------------- | -------------- |
| Type    | Simple filtering         | Condition-based filtering    | Visual filter  |
| Use     | Select values (checkbox) | Apply rules (>, <, contains) | Interactive UI |
| Example | Year = 2023              | Sales > 5000                 | Region buttons |

Key Point:

Basic = simple
Advanced = logical conditions
Slicer = user interaction

2. Using the Sales dataset, apply a Basic Filter to display only the sales data for the year 2023.

Select Basic Filtering
Choose Year = 2023

Result: Only records like 2023-01-10, 2023-07-18, etc. will be shown

4. What is a Visual-Level Filter, and how does it differ from Page-Level and Report-Level Filters?
-  | Type         | Scope                       |
| ------------ | --------------------------- |
| Visual-Level | Only one visual             |
| Page-Level   | All visuals on current page |
| Report-Level | Entire report               |

Example:

Visual filter → Only one chart shows Electronics
Page filter → All visuals on page show Electronics
Report filter → Entire report shows Electronics

6. What is the role of Drill-Down in Power BI? How does it improve data exploration?
